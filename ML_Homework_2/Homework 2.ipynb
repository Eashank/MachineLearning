```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("iris.data")
data.columns = ["sep-len","sep-wid","ped-len","ped-wid","class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.iloc[:, 0:4].values,data.iloc[:, 4].values, test_size = 0.2, random_state=0)
```

```python
standard = StandardScaler()
x_train = standard.fit_transform(x_train)
x_test = standard.transform(x_test)
x_test
```

```python
analysis = LinearDiscriminantAnalysis()
x_train = analysis.fit_transform(x_train,y_train)
x_test = analysis.transform(x_test)
```

```python
classifier = RandomForestClassifier(random_state=0)

classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pred))

#Since the accurace score is quite high I think the results of the LDA was good.
```
